```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have data or API access to both Spotify and Apple Music

# Sample data
spotify_data = {
    'popularity': [70, 80, 90, 85, 60],
    'duration_ms': [200000, 190000, 180000, 195000, 210000]
}

apple_music_data = {
    'popularity': [65, 75, 85, 80, 55],
    'duration_ms': [198000, 192000, 185000, 197000, 215000]
}

# Create dataframes
spotify_df = pd.DataFrame(spotify_data)
apple_music_df = pd.DataFrame(apple_music_data)

# Summary statistics
print("Summary statistics for Spotify:")
print(spotify_df.describe())

print("\nSummary statistics for Apple Music:")
print(apple_music_df.describe())

# Visualization
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(spotify_df['popularity'], bins=5, alpha=0.7, color='green', label='Spotify')
plt.hist(apple_music_df['popularity'], bins=5, alpha=0.7, color='blue', label='Apple Music')
plt.title('Popularity Distribution')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(spotify_df['duration_ms'], spotify_df['popularity'], color='green', label='Spotify')
plt.scatter(apple_music_df['duration_ms'], apple_music_df['popularity'], color='blue', label='Apple Music')
plt.title('Duration vs Popularity')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.legend()

plt.tight_layout()
plt.show()
```
